```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib import colors
from matplotlib.lines import Line2D
import matplotlib.gridspec as gridspec
from matplotlib.colors import LogNorm
from matplotlib.ticker import LogFormatterMathtext
plt.style.use("paper.mplstyle")

from controls import chs, ms, mcfiles
from calc_sens import SensCalc
from controls import sd_ann_rate_dict as sd_anns
from controls import si_ann_rate_dict as si_anns
```

```python
def calc_sens(sigflux, syst_set):
    sc = SensCalc(sigflux)
    for mcpath in mcfiles:
        if syst_set in mcpath:
            sc.add_mus(mcpath)
            sc.add_mub(mcpath)
    sc.set_sens()
    return sc.sens
```

```python
lw=3
cs = ["#C843C6", "#00B8C8", "#1B2984"]
desc_strs = [r'$b\bar{b}$', r'$W^{+}W^{-}$', r'$\tau^{+}\tau^{-}$']

syst_sets = ['HoleIce1', 'Nominal', 'HoleIce3', 'HoleIce4', 'HoleIce5', 'BulkIce1', 'BulkIce2', 'BulkIce5']
for syst_set in syst_sets:
    fig=plt.figure(figsize=(10,10))
    legend_elements = []
    for ds,c,ch in zip(desc_strs,cs,chs):
        sens = []
        good_ms = []
        for m in ms:
            try:
                sens.append(calc_sens('ch%d-m%d' % (ch,m), syst_set))
                good_ms.append(m)
            except:
                print('ch%d-m%d')
        legend_elements.append(Line2D([0], [0], color=c, label=ds, lw=lw))
        plt.plot(good_ms, sens, color=c, lw=lw)
    for c,l in zip(cs,tyl):
        plt.plot(l[0],l[1],'--', color=c, lw=lw)

    legend_elements.append(Line2D([0], [0], color='black', linestyle='--', label='Previous IceCube limits', lw=lw))
    legend_elements.append(Line2D([0], [0], color='black', label='Analysis sensitivity', lw=lw))

    plt.loglog()
    plt.xlim(500,10000)
    plt.xlabel(r'$m_{\chi}$ (GeV)',fontsize=32)
    plt.ylabel(r'$\sigma_{\chi N}^{\rm{SD}}$ (cm$^2$)',fontsize=32)
    plt.legend(handles=legend_elements, loc='upper left', ncol=2, prop={"size":23}, frameon=False)
    plt.tick_params(labelsize=32)
    plt.title(syst_set)
    plt.grid()
    plt.savefig('plots/%s_sensitivity.pdf' % syst_set)
    plt.show()
```
