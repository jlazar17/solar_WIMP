```python
ws_chan_dict = {5:"bb", 8:"WW", 11:"tautau"}
```

```python
def get_ws_sunsfc(ch, m):
    ws_fluxes = np.genfromtxt("/data/user/qliu/DM/wimpsim-4.1.2/scr/results_ann/dat-wa/wa-m%d-ch%d-sun-sum.dat" % (m, ch))
    return ws_fluxes[8], ws_fluxes[9]
```

```python
plt.plot(np.load("data/sunsfc_dn_dz/ch5_m1000_pythia_sunsfc.npy")[0])
plt.plot(get_ws_sunsfc(5, 1000)[0])
plt.semilogy()
plt.show()
```

```python
def comparison_plot(m):
    for ch in [5, 8, 11]:
        ws_nu, ws_nubar = get_ws_sunsfc(ch,m)
        _ = np.load("data/sunsfc_dn_dz/ch%d_m%d_pythia_sunsfc.npy" % (ch, m))
        qr_nu, qr_nubar = _[0], _[1]
        plt.plot(ws_nu, "r", label="WIMPSim")
        plt.plot(ws_nubar, "r--")
        plt.plot(qr_nu, "b", label="Pythia")
        plt.plot(qr_nubar, "b--")
        plt.semilogy()
        plt.legend(frameon=False)
        plt.title("ch%d, m%d" % (ch, m))
        plt.show()
```

```python
comparison_plot(1000)
```

```python
comparison_plot(5000)
```
