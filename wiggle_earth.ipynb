```python
ang_err = np.genfromtxt('data/3y_solar_WIMP_ang_res.csv', delimiter=',').T
```

```python
plt.plot(ang_err[0], ang_err[1])
plt.semilogx()
plt.show()
```

```python
tck = splrep(np.log(ang_err[0]), ang_err[1])
plt.plot(ang_err[0], ang_err[1])
ee = np.logspace(np.log10(ang_err[0][0]), np.log10(ang_err[0][-1]))
plt.plot(ee, [splev(np.log(e),tck) for e in ee])
plt.semilogx()
plt.show()
```

```python
np.save('data/3y_ang_error_spline.npy', tck, allow_pickle=True)
```

```python
maybe = np.load('data/3y_ang_error_spline.npy', allow_pickle=True)
plt.plot(ang_err[0], ang_err[1])

ee = np.logspace(np.log10(ang_err[0][0]), np.log10(ang_err[0][-1]))
plt.plot(ee, [splev(np.log(e),maybe) for e in ee])
plt.semilogx()
plt.show()
```
